# **🟢Project Description: Stock Portfolio Tracker**


The Stock Portfolio Tracker is a simple Python-based application that calculates the total investment value of stocks selected by a user. The program allows users to enter stock names and quantities, then computes the total investment based on predefined stock prices stored in a dictionary.

This project demonstrates the use of basic Python concepts such as dictionaries, user input, loops, conditional statements, arithmetic operations, and optional file handling.

The system first displays available stocks and their predefined prices. The user then inputs the number of stocks they want to invest in, followed by the stock name and quantity. The program validates whether the entered stock exists in the predefined list. If valid, it calculates the investment for each stock by multiplying the price with the quantity and adds it to the total investment value.

Finally, the program displays a portfolio summary including stock name, quantity, price per share, individual investment value, and total investment amount. Additionally, the user has the option to save the portfolio details into a text file for record-keeping.

This project is designed to strengthen fundamental programming skills and understand how basic financial calculations can be automated using Python.

This project calculates how much money a person invested in stocks using predefined prices.

# **🟢Goal**



* Take stock names and quantities from user

* Use a hardcoded dictionary for stock prices

* Calculate total investment value

* Optionally save result to a .txt file

# **🟢Python Code**

```python
# Stock Portfolio Tracker

# Hardcoded stock prices dictionary
stock_prices = {
    "AAPL": 180,
    "TSLA": 250,
    "GOOG": 140,
    "MSFT": 320,
    "AMZN": 130
}

total_investment = 0
portfolio = {}

print("Available Stocks:", list(stock_prices.keys()))

# Number of stocks user wants to add
n = int(input("How many different stocks do you want to invest in? "))

for i in range(n):
    stock_name = input("Enter stock name: ").upper()

    if stock_name in stock_prices:
        quantity = int(input("Enter quantity: "))

        investment = stock_prices[stock_name] * quantity
        total_investment += investment

        portfolio[stock_name] = quantity

    else:
        print("Stock not available in price list!")

print("\n----- Portfolio Summary -----")

for stock, quantity in portfolio.items():
    price = stock_prices[stock]
    print(f"{stock} -> Quantity: {quantity}, Price: {price}, Total: {price * quantity}")

print("Total Investment Value:", total_investment)

# Optional: Save to file
save_option = input("\nDo you want to save this to a file? (yes/no): ").lower()

if save_option == "yes":
    with open("portfolio.txt", "w") as file:
        file.write("Portfolio Summary\n")
        file.write("------------------\n")
        for stock, quantity in portfolio.items():
            price = stock_prices[stock]
            file.write(f"{stock} -> Quantity: {quantity}, Price: {price}, Total: {price * quantity}\n")
        file.write(f"\nTotal Investment Value: {total_investment}")

    print("Portfolio saved to portfolio.txt successfully!")

print("\nThank you for using Stock Portfolio Tracker!")
```

# **🟢Upgrade the code**

```python
#     Stock Portfolio Tracker

import csv

# Hardcoded stock prices
stock_prices = {
    "AAPL": 180,
    "TSLA": 250,
    "GOOG": 140,
    "MSFT": 320,
    "AMZN": 130
}

portfolio = {}


# Add Stock Function

def add_stock():
    stock_name = input("Enter stock name: ").upper()

    if stock_name not in stock_prices:
        print("❌ Stock not available!")
        return

    try:
        quantity = int(input("Enter quantity: "))
        if quantity <= 0:
            print("❌ Quantity must be greater than 0!")
            return
    except ValueError:
        print("❌ Please enter a valid number!")
        return

    if stock_name in portfolio:
        portfolio[stock_name] += quantity
    else:
        portfolio[stock_name] = quantity

    print("✅ Stock added successfully!")


# Calculate Total Investment

def calculate_total():
    total = 0
    for stock, quantity in portfolio.items():
        total += stock_prices[stock] * quantity
    return total


# Display Portfolio

def display_portfolio():
    if not portfolio:
        print("⚠ Portfolio is empty!")
        return

    total = calculate_total()

    print("\n----- Portfolio Summary -----")
    print("Stock | Price | Quantity | Total | Percentage")
    print("------------------------------------------------")

    for stock, quantity in portfolio.items():
        price = stock_prices[stock]
        stock_total = price * quantity
        percentage = (stock_total / total) * 100

        print(f"{stock} | {price} | {quantity} | {stock_total} | {percentage:.2f}%")

    print("------------------------------------------------")
    print("Total Investment Value:", total)



# Save Portfolio to CSV

def save_to_csv():
    if not portfolio:
        print("⚠ Nothing to save!")
        return

    with open("portfolio.csv", "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Stock", "Price", "Quantity", "Total"])

        for stock, quantity in portfolio.items():
            price = stock_prices[stock]
            writer.writerow([stock, price, quantity, price * quantity])

    print("✅ Portfolio saved to portfolio.csv successfully!")


# Main Menu

def main():
    while True:
        print("\n===== Stock Portfolio Tracker =====")
        print("1. Add Stock")
        print("2. View Portfolio")
        print("3. Save Portfolio to CSV")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            add_stock()
        elif choice == "2":
            display_portfolio()
        elif choice == "3":
            save_to_csv()
        elif choice == "4":
            print("Thank you for using Stock Portfolio Tracker!")
            break
        else:
            print("❌ Invalid choice! Please select 1-4.")


# Run the program
main()
```
