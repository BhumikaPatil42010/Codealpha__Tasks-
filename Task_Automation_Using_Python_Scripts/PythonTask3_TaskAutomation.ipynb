* Therefore, an automated solution is required to:

# **4. Proposed Solution**

A Python automation script is developed that reads a text file, detects email addresses using Regular Expressions (Regex), removes duplicate emails, and saves them into an output file automatically.

Additionally, the system generates a log file containing execution time and number of extracted emails.

# **6. Python Code :**

```python
import re
import os
from datetime import datetime
from colorama import Fore, init

init(autoreset=True)

print("Smart Email Extraction Tool Started...\n")

input_file = "input.txt"
output_file = "emails.txt"
log_file = "log.txt"

try:
    # Read file
    with open(input_file, "r") as file:
        content = file.read()

    # Extract emails
    emails = re.findall(
        r'[a-zA-Z0-9._%+-]+@[a-zA-Z0-9.-]+\.[a-zA-Z]+',
        content
    )

    # Remove duplicates
    unique_emails = set(emails)

    # Save emails
    with open(output_file, "w") as out:
        for email in unique_emails:
            out.write(email + "\n")

    # Logging
    with open(log_file, "a") as log:
        log.write(
            f"{datetime.now()} - Extracted {len(unique_emails)} emails\n"
        )

    print(f"SUCCESS: {len(unique_emails)} unique emails saved")
    print("Output file created: emails.txt")

except Exception as e:
    print("ERROR:", e)
```
