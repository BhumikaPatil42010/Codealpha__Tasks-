# **Python programming Task 1 : HANGMAN GAME**

# **1. Introduction**

Hangman is a simple text-based word guessing game.
In this game, the computer randomly selects a word from a predefined list, and the player has to guess the word one letter at a time.

The player is allowed only 6 incorrect guesses.
If the player guesses the word before 6 wrong attempts, they win. Otherwise, they lose.

# **2. How the Game Works (Step-by-Step)**

**Step 1: Word Selection**

A list of 5 predefined words is created.

Using random.choice(), one word is selected randomly.

**Step 2: Create Blank Display**

A list of underscores _ is created.

The number of underscores equals the length of the chosen word.

**Step 3: Game Loop (Main Logic)**

The game runs inside a while loop.

The loop continues until:

The player guesses the complete word
OR

The player makes 6 wrong guesses

**Step 4: Taking User Input**

The user enters one letter.

The input is converted to lowercase.

The program checks:

✔ If the letter is correct:

Replace the corresponding _ with the letter.

Display updated word.

❌ If the letter is wrong:

Increase incorrect guess count.

Reduce remaining attempts.

**Step 5: Win or Lose Condition**

After the loop ends:

If no _ remains → 🎉 Player Wins

If 6 incorrect guesses completed → 💀 Player Loses

# **3. CODE :**

```python
import random

# Step 1: Predefined word list (5 words)
words = ["python", "laptop", "engineer", "college", "project"]

# Step 2: Randomly choose a word
chosen_word = random.choice(words)

# Step 3: Create an empty display with underscores
display = []
for letter in chosen_word:
    display.append("_")

# Step 4: Game variables
incorrect_guesses = 0
max_attempts = 6
guessed_letters = []

print("🎮 Welcome to Hangman Game!")
print("You have 6 incorrect guesses allowed.")

# Step 5: Game Loop
while incorrect_guesses < max_attempts and "_" in display:

    print("\nCurrent Word: ", " ".join(display))
    guess = input("Enter a letter: ").lower()

    # Check if already guessed
    if guess in guessed_letters:
        print("You already guessed this letter!")
        continue

    guessed_letters.append(guess)

    # If guess is correct
    if guess in chosen_word:
        print("Correct guess! ✅")
        for i in range(len(chosen_word)):
            if chosen_word[i] == guess:
                display[i] = guess
    else:
        incorrect_guesses += 1
        print("Wrong guess! ❌")
        print("Remaining attempts:", max_attempts - incorrect_guesses)

# Step 6: Final Result
if "_" not in display:
    print("\n🎉 Congratulations! You won!")
else:
    print("\n💀 Game Over!")
    print("The correct word was:", chosen_word)
```
